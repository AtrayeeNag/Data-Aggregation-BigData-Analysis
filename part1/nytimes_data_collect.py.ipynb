```python
from nytimesarticle import articleAPI
from bs4 import BeautifulSoup
import time
import pandas as pd
import urllib.request
import numpy as np
```

```python
api = articleAPI('nGAzvObUhTbbqP7K3rfJdk7VhCID5EEK')
```

```python
def parse_articles(articles):
    news = []
    for i in articles['response']['docs']:
        dic = {}
        dic['id'] = i['_id']
        dic['url'] = i['web_url']
        news.append(dic)
    return(news)
```

```python
# key_words = ['Google', 'gmail', 'youtube', 'adsense']

# key_words = ['amazon', 'kindle', 'Alexa', 'Bezos', 'Audible', 'twitch', 'imdb']

#key_words = ['facebook', 'Instagram', 'WhatsApp', 'Oculus', 'Zuckerberg']

key_words = ['microsoft', 'bill gates', 'azure', 'skype', 'github', 'sharepoint', 'windows 7', 'internet explorer']


for word in key_words:
    try:
        articles = api.search( q = word, begin_date= 20190101 )

        hits = articles['response']['meta']['hits']
        loop_count = hits//10

        for j in range(0, loop_count-1):
            print(j)
            time.sleep(7)

            articles = api.search( q = word, begin_date= 20190101, page=j )
            article_info = parse_articles(articles)
            article_df = pd.DataFrame.from_dict(article_info, orient='columns')
            article_df.to_csv('data_files/nyt/nyt_3.csv' , encoding='utf-8', mode='a', header=False, index = False)
    except KeyError as e:
        print("KeyError occurred")
        continue
```
